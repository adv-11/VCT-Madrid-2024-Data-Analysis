```python
import seaborn as sns
team_stats = df.groupby('Team')['KD'].agg(['mean', 'var']).reset_index()

# Plot the graph
plt.figure(figsize=(10, 6))
sns.scatterplot(data=team_stats, x='var', y='mean', hue='Team', s=200)

# Add labels for each point
for i, team in enumerate(team_stats['Team']):
    plt.text(team_stats['var'][i], team_stats['mean'][i], team, fontsize=10)

# Set plot labels and title
plt.xlabel('Variance of KD')
plt.ylabel('Average of KD')
plt.title('Variance vs. Average of Player KD by Team')

# Show the plot
plt.grid(True)
plt.show()
```

```python
player_stats = df.groupby(['Team', 'Player'])['KD'].mean().reset_index()

# Calculate average KD for each team
team_avg_kd = player_stats.groupby('Team')['KD'].mean().sort_values(ascending=False)

# Sort player_stats based on average KD of teams
player_stats_sorted = player_stats.set_index('Team').loc[team_avg_kd.index].reset_index()

# Plot the bar graph
plt.figure(figsize=(13, 7))
sns.barplot(data=player_stats_sorted, x='KD', y='Team', color='darkblue')


# Plot the dots for player KD values
sns.scatterplot(data=player_stats_sorted, x='KD', y='Team', s=100, color='orange')

#  Highlight highest and lowest KD players
for team in player_stats_sorted['Team'].unique():
    highest_kd_player = player_stats_sorted[player_stats_sorted['Team'] == team].nlargest(1, 'KD')
    lowest_kd_player = player_stats_sorted[player_stats_sorted['Team'] == team].nsmallest(1, 'KD')

    plt.scatter(highest_kd_player['KD'], highest_kd_player['Team'], color='orange', marker='o', s=200, edgecolors='black')

    plt.scatter(lowest_kd_player['KD'], lowest_kd_player['Team'], color='orange', marker='o', s=200, edgecolors='black')

    plt.text(highest_kd_player['KD'] , highest_kd_player['Team'], highest_kd_player['Player'].values[0], fontsize=10, ha='right')

    plt.text(lowest_kd_player['KD'] , lowest_kd_player['Team'], lowest_kd_player['Player'].values[0], fontsize=10, ha='right')



# Set plot labels and title
plt.xlabel('Average KD')
plt.ylabel('Team')
plt.title('Average KD of Players by Team')

# Show the plot

plt.show()
```

```python
# Graph 1: Bar plot of KD by Player
plt.figure(figsize=(10, 6))
sns.barplot(x='ADR', y='Player', data=df.sort_values(by='KD'))
plt.title('ADR by Player')
plt.xlabel('ADR')
plt.ylabel('Player')
plt.show()
```

```python
# Scatter plot of FKPR and FDPR colored by Team
plt.figure(figsize=(10, 6))
sns.scatterplot(x='FKPR', y='FDPR', data=df, hue='Team', s=100)

# Add labels for player names
for i in range(len(df)):
    plt.text(df['FKPR'][i], df['FDPR'][i], df['Player'][i], fontsize=8, ha='center', va='bottom')

plt.title('FKPR vs FDPR by Player')
plt.xlabel('FKPR (First Kill per Round)')
plt.ylabel('FDPR (First Death per Round)')
plt.legend(title='Team')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='KAST%', y='Team', data=df)
plt.title('KAST by Team')
plt.xlabel('KAST')
plt.ylabel('Team')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Team', y='ACS', data=df, hue='Team')
plt.title('ACS vs Team')
plt.xlabel('Rounds')
plt.ylabel('ACS')
plt.legend(title='Team')
plt.show()
```
